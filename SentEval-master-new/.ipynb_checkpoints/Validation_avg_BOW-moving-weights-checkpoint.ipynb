```python
from __future__ import absolute_import, division
#from sklearn.decomposition import TruncatedSVD

#import Model
import numpy as np 
from keras.preprocessing.sequence import pad_sequences
from nltk.corpus import stopwords
import senteval
import pickle
import copy
import operator
import nltk
import Model

import torch
#from pytorch_pretrained_bert import GPT2Tokenizer, GPT2Model, GPT2LMHeadModel

PATH_SENTEVAL = '../'
PATH_TO_DATA = 'data'

maxLength = 30    
wordDim = 300    # 300 
num_neurons = [2*wordDim,2*wordDim]   # [600,600f]
extractLayer = ['X_input_first','output_cloze']
kernel_reg = 0.001
batch_size = 400
merge_mode = 'concat'

#checkpoint = "../checkpoint_own_data/Normal-loss-0.5497-cata_acc-0.7909-val_loss-0.7405-cate_acc_val-0.7029.hdf5"  # Change 
checkpoint = "../cloze_checkpoint/Middle-Cloze-loss-0.0643-val_loss-0.0654.hdf5"
processedPath = "../processed/"
tokenizerFile = 'tokenizer.txt'
embeddingFile =  'embedding_matrix.txt'


embeddingPath = "../Embedding/glove.840B.300d.txt"  # Change 
dicFile = 'wordDic.txt'
marked_token = "maskedtoken"
tf_idf_file = "tf-idf-dic.txt"

n_component = 5

deepLearningModel = False

# GPT2 
# tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
# model = GPT2Model.from_pretrained('gpt2')

# Glove
loadGloveEmbedding = False
count = 0
model = 0
NonExiste ={}
POS_Dic = {}

if loadGloveEmbedding:
    print("Load embedding")
    embeddings_index = dict()
    f = open(embeddingPath  , 'r', errors = 'ignore', encoding='utf8')  # "word2vec.6B.50d.txt" for test 
    for line in f:
        splitLine = line.split(' ')  # For 3B  splitLine = line.split()
        word = splitLine[0]
        coefs = np.array([float(val) for val in splitLine[1:]])
        embeddings_index[word] = coefs
    f.close()
    print("Finish loading")
    loadGloveEmbedding = False
```

```python
with open(processedPath + tf_idf_file , "rb") as input_file:
    tf_idf_val = pickle.load(input_file)
stop_words = set(stopwords.words('english')) 

with open(processedPath + tokenizerFile , "rb") as input_file:
    t = pickle.load(input_file)

with open(processedPath + embeddingFile , "rb") as input_file:
    embedding_matrix = pickle.load(input_file)

```

```python
# SentEval prepare and batcher
def prepare(params, samples):
    return

def batcher(params, batch):
    global embeddings_index
    global count
    global model
    sentences = [' '.join(s) for s in batch]
        
    sentenceList = []
    pos_output= []
    for sentence in sentences:
        
        totalVec = []
        sentence = sentence.replace("n 't", "n't")
        sentence = sentence.replace("-"," ")
        #print(sentence)
        sentence =  nltk.word_tokenize(sentence)
        
        
        gloveVec = []
        tf_avg = []
        for vocab in sentence:
            try: 
                if vocab in tf_idf_val.keys():
                    gloveVec.append(embeddings_index[vocab]*tf_idf_val[vocab]/5)
                else:
                    gloveVec.append(embeddings_index[vocab]*tf_idf_val[vocab])
            except:
                gloveVec.append(np.zeros(300))
#                 if vocab not in NonExiste.keys:
#                     NonExiste[vocab] = 1
#                 else: 
#                     NonExiste[vocab] += 1 
        #print(len(sentence))
        if len(gloveVec) == 0:
            tf_avg = np.zeros([1,300])[0]
        else:    
            tf_avg = np.mean(np.array(gloveVec),axis = 0 , keepdims = True)[0]
        



#         if len(gloveVec) != 0:
#             for i,val in enumerate(gloveVec):
    
#                 if i == 0 and len(gloveVec) > 1:
#                     gloveVec[i] = 0.5*gloveVec[i] + 0.5*gloveVec[i+1]
#                 elif i == len(gloveVec)-1 and i != 0:
#                     gloveVec[i] = 0.5*gloveVec[i] + 0.5*gloveVec[i-1]
#                 else:
#                     if len(gloveVec) > 2:
#                         #gloveVec[i] = 0.5*gloveVec[i] + 0.25*gloveVec[i-1] + 0.25*gloveVec[i+1]
#                         gloveVec[i] = 0.5*gloveVec[i-1] + 0.5*gloveVec[i+1]
            
        stopWord = []
        nonStopWord = []
        for i,vocab in enumerate(sentence):
            lower = vocab.lower() 
            if lower in stop_words:  # has to be lower case 
                stopWord.append(gloveVec[i])         
            else:
                nonStopWord.append(gloveVec[i])
               
        #print("Stopword:{}, nonStopword:{}, total: {}".format(len(stopWord),len(nonStopWord),len(sentence)))
        if len(stopWord) == 0:
            stopWord = np.zeros([1,300])
        if len(nonStopWord) == 0:
            nonStopWord = np.zeros([1,300])

        average = np.mean(np.array(list(stopWord) + list(nonStopWord)),axis = 0 , keepdims = True)
        stopWord = np.mean(np.array(stopWord),axis= 0, keepdims = True)     
        nonStopWord = np.mean(np.array(nonStopWord),axis= 0, keepdims = True)
#         min_ = np.amin(np.array(list(stopWord) + list(nonStopWord)), axis=0,keepdims = True)
#         max_ = np.amax(np.array(list(stopWord) + list(nonStopWord)), axis=0,keepdims = True)
        output = (list(average[0]) + list(stopWord[0]) + list(nonStopWord[0])+ list(tf_avg))
        print(len(output))
#         print(len(output)) 
        sentenceList.append(output)
    
    return sentenceList
    
    #output = np.concatenate((np.array(sentenceList),np.array(pos_output)),axis = 1)
    

#     Combine deep learning model
    if deepLearningModel:
        for i,sentence in enumerate(sentences):
            
            sentence = sentence.replace("n 't", "n't")
            sentence = sentence.replace("-"," ")
            #print(sentence)
            sentences[i] = nltk.word_tokenize(sentence)
            
        sentences = t.texts_to_sequences(sentences)
        sentences = pad_sequences(sentences, maxlen=maxLength, padding='post')

        if count == 0:
            model = Model.buildClozeModel(sentences,kernel_reg,num_neurons,merge_mode, len(embedding_matrix),maxLength, wordDim,embedding_matrix)
            #print(model.summary())
            print("load weights")
            model.load_weights(checkpoint)
            count += 1

        # Extract th output 

        output = Model.extractClozeHiddenState(extractLayer,model,sentences)
        output = np.concatenate((output[0],output[1]),axis = 1)
        
#         output = model.predict(sentences)
#         output = np.concatenate((output[0],output[1]),axis = 1)
    
#         output = np.concatenate((output,output1),axis = 1)
        print(output.shape)
        return output
        
#         sentenceList = np.concatenate((output,output1),axis = 1)
    
    return sentenceList
```

```python
tasks =  ['MR','SUBJ','CR','MPQA','TREC','MRPC']
result = []
for task in tasks:
    print(task)
    
    params_senteval = {'task_path': PATH_TO_DATA, 'usepytorch': True, 'kfold': 10}
    params_senteval['classifier'] = {'nhid': 0, 'optim': 'adam', 'batch_size': 512,
                                     'tenacity': 5, 'epoch_size': 4}
    # Set up logger
    se = senteval.engine.SE(params_senteval, batcher, prepare)
    transfer_tasks =    [task]  # ['CR','MPQA','TREC','MRPC']
    # 'MR', 'CR' ,'MPQA', 'SUBJ','TREC', 'MRPC','SICKEntailment', 'SICKRelatedness'  ,'TREC','MRPC'
    results = se.eval(transfer_tasks)
    print(results)
    result.append(results)
```
